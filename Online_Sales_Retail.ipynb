# Introduction

## Dataset Description

This dataset provides a comprehensive overview of online sales transactions across different product categories. Each row represents a single transaction with detailed information such as the order ID, date, category, product name, quantity sold, unit price, total price, region, and payment method.

## Atribute Information

**Attribute Information:**

- **Order ID:** Unique identifier for each sales order.
- **Date:** Date of the sales transaction.
- **Category:** Broad category of the product sold (e.g., Electronics, Home Appliances, Clothing, Books, Beauty Products, Sports).
- **Product Name:** Specific name or model of the product sold.
- **Quantity:** Number of units of the product sold in the transaction.
- **Unit Price:** Price of one unit of the product.
- **Total Price:** Total revenue generated from the sales transaction (Quantity * Unit Price).
- **Region:** Geographic region where the transaction occurred (e.g., North America, Europe, Asia).
- **Payment Method:** Method used for payment (e.g., Credit Card, PayPal, Debit Card).


## Analysis Objectives

1. Analyze sales trends over time to identify seasonal patterns or growth opportunities.
2. Explore the popularity of different product categories across regions.
3. Investigate the impact of payment methods on sales volume or revenue.
4. Identify top-selling products within each category to optimize inventory and marketing strategies.
5. Evaluate the performance of specific products or categories in different regions to tailor marketing campaigns accordingly.

# Data Preparation

## Creating a Copy of the Original Dataframe

```python
# Create a copy of the original dataframe
df_original = df.copy()

# Use the copy of the original dataframe for data preparation
df_preparation = df_original.copy()
```

## Converting Columns to Appropriate Data Types:

```python
# Convert Date
df_preparation['Date'] = pd.to_datetime(df_preparation['Date'])
```

## Feature Engineering

```python
# Adding Day of Year column
df_preparation['Year'] = df_preparation['Date'].dt.year

# Adding Day of Month column
df_preparation['Month'] = df_preparation['Date'].dt.month

# Adding Day of Week column
df_preparation['Day of Week'] = df_preparation['Date'].dt.day_name()

# Create a new column to define discount or price per unit
df_preparation['Discounted'] = df_preparation['Total Revenue'] < df_preparation['Units Sold'] * df_preparation['Unit Price']
```

## Sorting and Indexing Data

```python
# Sort by date and set the index
df_preparation = df_preparation.sort_values(by='Date').set_index('Date')
```

## Ensuring Category Codes are Consistent:

```python
# Ensuring Category Codes are Consistent
# Check unique values in category columns
print(df_preparation['Product Category'].unique())
print(df_preparation['Payment Method'].unique())
print(df_preparation['Region'].unique())

# Correct inconsistencies (if any)
# df_preparation['Product Category'] = df_preparation['Product Category'].replace({'home appliances': 'Home Appliances'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
import calendar
from pandas.plotting import parallel_coordinates

# Set the overall theme for the visualizations
sns.set_theme(style="whitegrid")
```

## Visualization Using Day of Week Column

```python
# Grouping data by day of the week and calculating total sales
day_sales = df_preparation.groupby('Day of Week').agg({'Total Revenue': 'sum'}).reset_index()

# Sorting days of the week in proper order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_sales['Day of Week'] = pd.Categorical(day_sales['Day of Week'], categories=weekday_order, ordered=True)
day_sales = day_sales.sort_values('Day of Week')

# Plotting sales by day of the week
plt.figure(figsize=(14, 7))
sns.barplot(x='Day of Week', y='Total Revenue', data=day_sales, palette="viridis", order=weekday_order, hue='Day of Week', dodge=False, legend=False)
plt.title('Sales Analysis by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
for index, value in enumerate(day_sales['Total Revenue']):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

## 1 - Analyze sales trends over time to identify seasonal patterns or growth opportunities:

### Monthly Sales Trend Analysis

```python
# Group by month and year, and aggregate total revenue
monthly_sales = df_preparation.groupby(['Year', 'Month']).agg({'Total Revenue': 'sum'}).reset_index()

# Convert month numbers to month names for better readability in the plot
monthly_sales['Month'] = monthly_sales['Month'].apply(lambda x: calendar.month_abbr[x])

# Prepare a bright color palette
palette = sns.color_palette("husl", len(monthly_sales['Year'].unique()))

# Plotting monthly sales trend
plt.figure(figsize=(14, 7))
sns.lineplot(x='Month', y='Total Revenue', hue='Year', data=monthly_sales, palette=palette, marker='o', linewidth=2.5)
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)

# Annotation for the month with highest sales
max_sales_idx = monthly_sales['Total Revenue'].idxmax()
max_sales_month = monthly_sales.loc[max_sales_idx, 'Month']
max_sales_value = monthly_sales['Total Revenue'].max()
plt.annotate(f'Highest Sales\n{max_sales_month}: {max_sales_value:.2f}',
             xy=(max_sales_month, max_sales_value),
             xytext=(max_sales_idx % 12 + 2, max_sales_value + 500),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12, ha='center')

plt.legend(title='Year', title_fontsize='13', fontsize='11', loc='upper right', frameon=True)
plt.tight_layout()
plt.show()
```

## 2 - Explore the popularity of different product categories across regions:

### Sales Analysis by Product Category

```python
# Group by product category and aggregate total revenue
category_sales = df_preparation.groupby('Product Category').agg({'Total Revenue': 'sum'}).reset_index()

# Plotting sales by product category
plt.figure(figsize=(14, 7))
sns.barplot(x='Product Category', y='Total Revenue', data=category_sales, palette="viridis", hue='Product Category', dodge=False, legend=False)
plt.title('Sales by Product Category', fontsize=16)
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
for index, value in enumerate(category_sales['Total Revenue']):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Group by region and aggregate total revenue
region_sales = df_preparation.groupby('Region').agg({'Total Revenue': 'sum'}).reset_index()

# Plotting sales by region
plt.figure(figsize=(14, 7))
sns.barplot(x='Region', y='Total Revenue', data=region_sales, palette="Blues_d", hue='Region', dodge=False, legend=False)
plt.title('Sales by Region', fontsize=16)
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
for index, value in enumerate(region_sales['Total Revenue']):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

## 3 - Investigate the impact of payment methods on sales volume or revenue:

### Sales Analysis by Payment Method

```python
# Group by payment method and aggregate total revenue
payment_sales = df_preparation.groupby('Payment Method').agg({'Total Revenue': 'sum'}).reset_index()

# Plotting sales by payment method
plt.figure(figsize=(14, 7))
sns.barplot(x='Payment Method', y='Total Revenue', data=payment_sales, palette="cubehelix", hue='Payment Method', dodge=False, legend=False)
plt.title('Sales by Payment Method', fontsize=16)
plt.xlabel('Payment Method')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
for index, value in enumerate(payment_sales['Total Revenue']):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

## 4 - Identify top-selling products within each category to optimize inventory and marketing strategies:

### Identifying Top Selling Products

```python
# Group by product name and aggregate total revenue
top_products = df_preparation.groupby('Product Name').agg({'Total Revenue': 'sum'}).reset_index().sort_values(by='Total Revenue', ascending=False).head(10)

# Prepare a bright color palette
palette = sns.color_palette("magma", len(top_products['Product Name'].unique()))

# Plotting top selling products
plt.figure(figsize=(14, 7))
sns.barplot(x='Total Revenue', y='Product Name', data=top_products, palette=palette, hue='Product Name', orient='h', dodge=False, legend=False)
plt.title('Top Selling Products', fontsize=16)
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')
for index, value in enumerate(top_products['Total Revenue']):
    plt.text(value, index, f'{value:.0f}', ha='left', va='center', fontsize=10)
plt.tight_layout()
plt.show()
```

## 5 - Evaluate the performance of specific products or categories in different regions to tailor marketing campaigns accordingly

### Stacked Bar Plot for Sales per Product Category in Each Region:

```python
# Grouping data by Region and Product Category and aggregating total revenue
region_category_sales = df_preparation.groupby(['Region', 'Product Category']).agg({'Total Revenue': 'sum'}).reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(x='Region', y='Total Revenue', hue='Product Category', data=region_category_sales, palette='tab10')
plt.title('Total Revenue by Product Category and Region', fontsize=16)
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Heatmap to Display Total Revenue Data:

```python
# Create a pivot table for heatmap visualization
heatmap_data = region_category_sales.pivot(index='Region', columns='Product Category', values='Total Revenue')
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=.5)
plt.title('Heatmap: Total Revenue by Region and Product Category', fontsize=16)
plt.xlabel('Product Category')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

### Comparative Analysis among Regions:

```python
# Comparing the performance of specific products or categories in different regions
plt.figure(figsize=(14, 7))
sns.barplot(x='Region', y='Total Revenue', hue='Product Category', data=region_category_sales, dodge=True, palette='tab10')
plt.title('Total Revenue Comparison by Product Category in Different Regions', fontsize=16)
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Correlation Analysis

```python
# Calculating the correlation matrix
correlation_matrix = df_preparation[['Units Sold', 'Unit Price', 'Total Revenue']].corr()

# Plotting heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', annot_kws={"size": 12})
plt.title('Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.show()
```

## Heatmap of Sales by Region and Payment Method

```python
# Group by Region and Payment Method, and aggregate total revenue
heatmap_data = df_preparation.groupby(['Region', 'Payment Method']).agg({'Total Revenue': 'sum'}).unstack()

# Plotting heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=.5, annot_kws={"size": 10})  # custom font size
plt.title('Heatmap: Total Revenue by Region and Payment Method', fontsize=16)
plt.xlabel('Payment Method')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

## Parallels Coordinates Plot

```python
# Choosing a subset of data for parallel coordinates visualization
subset = df_preparation[['Units Sold', 'Unit Price', 'Total Revenue', 'Product Category']]
# Selecting 5 product categories for clearer visualization
subset = subset[subset['Product Category'].isin(subset['Product Category'].unique()[:5])]

# Plotting parallel coordinates
plt.figure(figsize=(14, 7))
parallel_coordinates(subset, 'Product Category', colormap=plt.get_cmap("Set1"))
plt.title('Parallel Coordinates Plot for Product Categories', fontsize=16)
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Pair Plot to View Relationships Between Variables

## Treemap for Sales by Product Category

```python
# Calculating total revenue for each product category
category_revenue = df_preparation.groupby('Product Category').agg({'Total Revenue': 'sum'}).reset_index()

# Plotting treemap
plt.figure(figsize=(14, 7))
squarify.plot(sizes=category_revenue['Total Revenue'], label=category_revenue['Product Category'], alpha=.8, color=sns.color_palette('Paired', len(category_revenue['Product Category'])))
plt.title('Treemap of Total Revenue by Product Category', fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```

## Pie Chart: Number of Transactions by Payment Method

```python
# Counting the number of transactions for each payment method
payment_method_counts = df_preparation['Payment Method'].value_counts()

# Plotting pie chart for payment methods
plt.figure(figsize=(14, 7))
colors = sns.color_palette("Accent", len(payment_method_counts))

# Generate the pie chart and capture the return values for customization
wedges, texts, autotexts = plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)

plt.title('Pie Chart: Number of Transactions by Payment Method', fontsize=16)

# Customize the annotations
for i, autotext in enumerate(autotexts):
    autotext.set_text(f"{payment_method_counts[i]} ({autotext.get_text()})")
    autotext.set_color('white')
    autotext.set_fontsize(14)
    autotext.set_weight('bold')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
